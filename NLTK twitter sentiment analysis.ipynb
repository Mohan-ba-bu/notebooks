```python
# remove single digit words
text3=[]
for i in range(len(text2)):
    temp=[]
    for j in text2[i].split():
        if len(j.strip())==1:
            continue
        else:
            temp.append(j)
    text3.append(' '.join(temp))
```

```python
# removing hyperlinks
import re
comp = re.sub(r'https://\S+|www.\s+','',a)
comp
```

```python
text4 =[]
for i in range(len(text3)):
    # removing hyperlinks
    temp = re.sub(r'https://\S+|www\.\S+','',text3[i])
    # removing digits/numbers
    temp = re.sub(r'\d+','',temp)
    if (len(temp)==1) | (len(temp)==0):
        continue
    else:
        text4.append(temp.replace('  ',' ').strip())
```

```python
text4
```

```python
text5 = []
for i in range(len(text4)):
    text5.append(text4[i].translate(str.maketrans('','',string.punctuation)))
```

```python
positive =0
negative = 0
neutral =0
pos_tweet=[]
neg_tweet = []
neu_tweet =[]
for i in text5:
    score = SentimentIntensityAnalyzer().polarity_scores(i)
    neg = score['neg']
    pos = score['pos']
    new = score['neu']
    if pos>neg:
        pos_tweet.append(i)
        positive+=1
    elif neg>pos:
        neg_tweet.append(i)
        negative+=1
    elif neg == pos:
        neutral+=1
        neu_tweet.append(i)
```

```python
def percentage(no,total):
    return round(100*(no/total),2)
```

```python
print('percentage of positive tweets', percentage(positive,len(text5)),'%')
print('percentage of negative tweets', percentage(negative,len(text5)),'%')
print('percentage of neutral tweets', percentage(neutral,len(text5)),'%')
```

```python
from PIL import Image
from wordcloud import WordCloud,STOPWORDS
```

```python
def create_wordcloud(text):
    mask = np.array(Image.open(r"cat.jpg"))
    stopwords = set(STOPWORDS)
    wc = WordCloud(background_color='white',
                  mask = mask,max_words=1000,
                   stopwords=stopwords,
                  repeat=True)
    wc.generate(str(text))
    wc.to_file('wc.png')
    print('Word Cloud Saved Successfully')
    path='wc.png'
    display(Image.open(path))
```
